```python
# y = 2*x + 3
# x = [1,2,3,4,5]
# y = [5,7,9,11,13]
# mean square error is called the cost function
# gradient descent is an algorithm that finds best fit line for given training data set
# graph is like bowl
```

```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_current = 0 
    b_current = 0
    iterations = 1000
    n = len(x)
    learning_rate = 0.08
    for i in range(iterations):
        y_predicted = m_current * x +  b_current
        # cost is mse 
        cost = (1/n) * sum([val ** 2 for val in (y-y_predicted)]) # use of list comprehensions val**2 --> square
        md = -(2/n) * sum(x * (y-y_predicted)) # partial derivate of m
        bd = -(2/n) * sum((y-y_predicted))     # partial derivate of b
        m_current = m_current - learning_rate * md
        b_current = b_current - learning_rate * bd
        print("m = {} , b = {} , cost = {}  , iteration = {}".format(m_current,b_current,cost,i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
gradient_descent(x,y)
```
